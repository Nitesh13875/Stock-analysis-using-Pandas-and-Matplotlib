```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('ADANIENT.NS.csv')
```

```python
company_data = data[['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
```

```python
company_data['Daily Return'] = company_data['Close'].pct_change()
```

```python
company='Adani Enterprises Limited'
start_date = '2023-01-01'
end_date = '2023-02-17'
period = company_data.loc[(company_data['Date']>= start_date) & (company_data['Date']<= end_date)]
```

```python
total_return = (period['Daily Return'] + 1).prod() - 1
if total_return<=0:
    Total_Loss=total_return
    print(f'Total_Loss:{Total_Loss}')
else:
    Total_Profit=total_return
    print(f'Total_Profit:{Total_Loss}')
```

```python
print(f"Company: {company}")
print(f"Start Date: {start_date}")
print(f"End Date: {end_date}")
print(f"Total Return: {total_return:.2%}")
```

```python
plt.figure(figsize=(10,7))
plt.xticks(rotation=45)
plt.plot(period['Date'], period['Daily Return'])
plt.title(f'Daily Returns for Adani Enterprises Limited from {start_date} to {end_date}')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.show()
```
